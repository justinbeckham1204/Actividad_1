```python
# Crear la base de datos (si no existe)
spark.sql("CREATE DATABASE IF NOT EXISTS bigdata2025")
```

```python
# (cargado de datos) Cargar el dataset que ya está en el catálogo Workspace
df = spark.table("workspace.bigdata2025.playstationdataset2025")

# Mostrar las primeras filas
display(df.limit(10))
```

```python
%sql
--evidencias de la carga con un SELECT * LIMIT 5.
USE BigData2025;

SELECT * FROM playstationdataset2025
LIMIT 5;
```

```python

%sql
--CONTEO DE REGISTROS
USE bigdata2025;

SELECT COUNT(*) AS total_registros
FROM playstationdataset2025;
```

```python
%sql
USE CATALOG workspace;
USE bigdata2025;
-- DESCRIBE NOMBRES Y TIPOS DE COLUMNAS
DESCRIBE TABLE playstationdataset2025;
```

```python
%sql
USE CATALOG workspace;
USE bigdata2025;
--CONSULTA CON FILTRO
SELECT *
FROM playstationdataset2025
WHERE Publisher = 'Rockstar Games'
LIMIT 5;
```
